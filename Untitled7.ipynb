```python
train='/content/Vegetable Images/train'
val='/content/Vegetable Images/validation'
test='/content/Vegetable Images/test'
```

```python
train_ds = keras.utils.image_dataset_from_directory(
    directory = train,
    labels='inferred',
    label_mode = 'int',
    batch_size=32,
    image_size=(256,256)
)

validation_ds = keras.utils.image_dataset_from_directory(
    directory = val,
    labels='inferred',
    label_mode = 'int',
    batch_size=32,
    image_size=(256,256)
)
```

```python
num_of_classes=len(train_ds.class_names)
num_of_classes
```

```python
model = Sequential()

model.add(Conv2D(32, (3, 3), input_shape=(256, 256, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())

model.add(Dense(256, activation='relu'))
model.add(Dense(15, activation='sigmoid'))  # Assuming 15 classes

# Compile the model
```

```python
model.summary()
```

```python
keras.utils.plot_model(
	model,
	show_shapes = True,
	show_dtype = True,
	show_layer_activations = True
)
```

```python
model.compile(optimizer='adam',loss=tf.losses.SparseCategoricalCrossentropy(from_logits=True),metrics=['accuracy'])
```

```python
hist=model.fit(train_ds, epochs=10, validation_data=validation_ds)
```
